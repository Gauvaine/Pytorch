```python
# 标注该类别的数量
# ax.annotate函数用于在指定位置添加文本注释
# round() 函数用于将一个数字四舍五入为指定的小数位数
# p.get_x() 获取条形的左边缘的 x 坐标；p.get_y() 获取条形的下边缘的 y 坐标
# p.get_width() 获取条形的宽度；p.get_height() 获取条形的高度。
def barw(ax):     
    for p in ax.patches:
        val = p.get_width() # 获取该类别的数量
        x = p.get_x()+ p.get_width()
        y = p.get_y() + p.get_height()/2
        ax.annotate(round(val,2),(x,y))
        
# labels_dataframe['label'].value_counts()统计label列中每个类别出现的次数，并返回一个包含类别和对应次数的Series对象
# index属性来获取这个Series的索引，即按照类别出现的次数从高到低的顺序来排列条形
plt.figure(figsize = (15,30))
ax0 = sns.countplot(y = leaves_data['label'],order = leaves_data['label'].value_counts().index)
barw(ax0)
plt.show()
```

```python
def load_data(csv_path, img_path, valid_ratio=0.2, resize_height=256, resize_width=256, mode='train'):
    data_info = pd.read_csv(csv_path, header=None)
    data_len = len(data_info.index) - 1
    train_len = int(data_len * (1 - valid_ratio))
    
    data = []
    if mode=='train':
        train_image = np.asarray(data_info.iloc[1:train_len, 0])
        train_label = np.asarray(data_info.iloc[1:train_len, 1])
        for index in range(len(train_image)):
            single_image_name = train_image[index]
            img = Image.open(img_path + single_image_name)
            transform = transforms.Compose([
                transforms.Resize((224, 224)),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.ToTensor()
            ])
            img = transform(img)
            label = train_label[index]
            number_label = class_to_num[label]
            data.append((img, number_label))
        return data
    elif mode=='valid':
        valid_image = np.asarray(data_info.iloc[train_len:, 0])
        valid_label = np.asarray(data_info.iloc[train_len:, 1])
        for index in range(len(valid_image)):
            single_image_name = valid_image[index]
            img = Image.open(img_path + single_image_name)
            transform = transforms.Compose([
                transforms.Resize((224, 224)),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.ToTensor()
            ])
            img = transform(img)
            label = valid_label[index]
            number_label = class_to_num[label]
            data.append((img, number_label))
        return data
    elif mode=='test':
        test_image = np.asarray(data_info.iloc[1:, 0])
        for index in range(len(test_image)):
            single_image_name = test_image[index]
            img = Image.open(img_path + single_image_name)
            transform = transforms.Compose([
                transforms.Resize((224, 224)),
                transforms.ToTensor()
            ])
            img = transform(img)
            data.append(img)
        return data
    else:
        return data
```

```python
train_path = '../data/train.csv'
test_path = '../data/test.csv'
# csv文件中已经有images的路径了，因此这里只到上一级目录
img_path = '../data/'

train_dataset = load_data(train_path, img_path, mode='train')
val_dataset = load_data(train_path, img_path, mode='valid')
test_dataset = load_data(test_path, img_path, mode='test')
```

```python
# 定义data loader
# num_workers，使用x个进程读取数据
train_loader = torch.utils.data.DataLoader(dataset=train_dataset,batch_size=256, shuffle=False,num_workers=5)
val_loader = torch.utils.data.DataLoader(dataset=val_dataset,batch_size=256, shuffle=False,num_workers=5)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset,batch_size=256, shuffle=False,num_workers=5)     
```

```python
# 可视化样本
# 在ax.imshow()方法中，要求输入的图像数据的维度顺序通常是（高度，宽度，通道数）
def show_images(imgs, num_rows, num_cols, titles=None, scale=1.5): 
    figsize = (num_cols * scale, num_rows * scale)
    _, axes = d2l.plt.subplots(num_rows, num_cols, figsize=figsize)
    axes = axes.flatten() # 将二维数组展平，便于遍历
    for i, (ax, img) in enumerate(zip(axes, imgs)):
        if torch.is_tensor(img):
            # 图片张量  
            img = img.numpy()
            if img.ndim == 3 and img.shape[0] == 3:  # 如果是三通道图像，将通道维度移动到最后
                img = img.transpose((1, 2, 0))
            ax.imshow(img)
        else:
            # PIL图片
            ax.imshow(img)
        # 隐藏坐标轴
        ax.axes.get_xaxis().set_visible(False)
        ax.axes.get_yaxis().set_visible(False)
        if titles:
            ax.set_title(titles[i])
    return axes
```

```python
# 可视化样本
inputs, classes = next(iter(val_loader))
show_images(inputs, 2, 4, titles = [num_to_class[c.item()] for c in classes]);
```

```python
# CPU or GPU
def get_device():
    return 'cuda' if torch.cuda.is_available() else 'cpu'

device = get_device()
print(device)
```

```python
# lenet
net = nn.Sequential(
    # 第一个卷积层使用2个像素的填充，来补偿卷积核5x5导致的特征减少
    nn.Conv2d(3, 32, kernel_size=5, padding=2), nn.Sigmoid(),
    nn.AvgPool2d(kernel_size=2, stride=2),
    nn.Conv2d(32, 64, kernel_size=5), nn.Sigmoid(),
    nn.AvgPool2d(kernel_size=2, stride=2),
    nn.Flatten(),
    nn.Linear(64 * 54 * 54, 512), nn.Sigmoid(),
    nn.Linear(512, 256), nn.Sigmoid(),
    nn.Linear(256, 176))
```

```python
# 检查模型：通过打印每一层输出的形状
X = torch.rand(size=(1, 3, 224, 224), dtype=torch.float32)
for layer in net:
    X = layer(X)
    print(layer.__class__.__name__,'output shape: \t',X.shape)
```

```python
lr, num_epochs = 0.2, 6
d2l.train_ch6(net, train_loader, val_loader, num_epochs, lr, d2l.try_gpu())
```

```python
# resnet
def resnet_block(input_channels, num_channels, num_residuals,first_block=False):
    blk = []
    for i in range(num_residuals):
        # 第一个模块：两个残差块的通道数同输入通道数一致
        # 其余模块：第一个残差块通道数翻倍，高和宽减半；第二个残差块不变
        if i == 0 and not first_block:
            blk.append(d2l.Residual(input_channels, num_channels,use_1x1conv=True, strides=2))
        else:
            blk.append(d2l.Residual(num_channels, num_channels))
    return blk
```

```python
b1 = nn.Sequential(nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3),
                   nn.BatchNorm2d(64), nn.ReLU(),
                   nn.MaxPool2d(kernel_size=3, stride=2, padding=1))
# 每个模块有2个残差块
b2 = nn.Sequential(*resnet_block(64, 64, 2, first_block=True))
b3 = nn.Sequential(*resnet_block(64, 128, 2))
b4 = nn.Sequential(*resnet_block(128, 256, 2))
b5 = nn.Sequential(*resnet_block(256, 512, 2))
```

```python
# 最后与GoogLeNet一样，在ResNet中加入全局平均汇聚层，以及全连接层输出
net = nn.Sequential(b1, b2, b3, b4, b5,
                    nn.AdaptiveAvgPool2d((1,1)),
                    nn.Flatten(), nn.Linear(512, 176))
```

```python
# ResNet中不同模块的输入形状
X = torch.rand(size=(1, 3, 224, 224))
for layer in net:
    X = layer(X)
    print(layer.__class__.__name__,'output shape:\t', X.shape)
```

```python
lr, num_epochs = 0.2, 10
d2l.train_ch6(net, train_loader, val_loader, num_epochs, lr, d2l.try_gpu())
```

```python
saveFileName = './submission.csv'
```

```python
## 预测
net.eval()
predictions = []

for batch in tqdm(test_loader):
    imgs = batch
    with torch.no_grad():
        logits = net(imgs.to(device))
    predictions.extend(logits.argmax(dim=-1).cpu().numpy().tolist())

preds = []
for i in predictions:
    preds.append(num_to_class[i])

test_data = pd.read_csv(test_path)
test_data['label'] = pd.Series(preds)
submission = pd.concat([test_data['image'], test_data['label']], axis=1)
submission.to_csv(saveFileName, index=False)
print("Done!!!!!!!!!!!!!!!!!!!!!!!!!!!")
```
